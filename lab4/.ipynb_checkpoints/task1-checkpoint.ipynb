```python
import matplotlib.pyplot as plt
import numpy as np

eq1 = lambda x: (-149 + 10.2 * x) / 3.2
eq2 = lambda x: (83 + 5.8 * x) / 16
eq3 = lambda x: (234 - 10.3 * x) / 7.3


x = np.linspace(-10, 10, 400)

plt.figure(figsize=(15, 5))


plt.subplot(1, 3, 1)
plt.plot(x, eq1(x), color='red', linestyle='dashed', label='$10.2x_1 - 3.2x_2 = 149$')
plt.grid(True)
plt.title('10.2x_1 - 3.2x_2 = 149')

plt.subplot(1, 3, 2)
plt.plot(x, eq2(x), color='green', linestyle='dashdot', label='$-5.8x_1 + 16x_2 = 83$')
plt.grid(True)
plt.title('-5.8x_1 + 16x_2 = 83')

plt.subplot(1, 3, 3)
plt.plot(x, eq3(x), color='magenta', linestyle='dotted', label='$10.3x_1 + 7.3x_2 = 234$')
plt.grid(True)
plt.title('10.3x_1 + 7.3x_2 = 234')
```
