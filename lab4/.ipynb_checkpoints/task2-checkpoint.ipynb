```python
import matplotlib.pyplot as plt
import numpy as np

#  
eq1 = lambda x: (-149 + 10.2 * x) / 3.2
eq2 = lambda x: (83 + 5.8 * x) / 16
eq3 = lambda x: (234 - 10.3 * x) / 7.3

eq11 =  (-149 + 10.2 * x) / 3.2
eq22 =  (83 + 5.8 * x) / 16
eq33 =  (234 - 10.3 * x) / 7.3

#  x
x = np.linspace(-10, 20, 400)  #    x

#  
plt.figure(figsize=(8, 8), dpi=100)

#   
plt.plot(x, eq1(x), color='red', linestyle='dashed', label='$10.2x_1 - 3.2x_2 = 149$')
plt.plot(x, eq2(x), color='green', linestyle='dashdot', label='$-5.8x_1 + 16x_2 = 83$')
plt.plot(x, eq3(x), color='magenta', linestyle='dotted', label='$10.3x_1 + 7.3x_2 = 234$')

x2_min = max(np.min(eq11), np.min(eq22), np.min(eq33)) 

plt.fill_between(x, eq22, x2_min, color='gray', alpha=0.3)
plt.fill_between(x, eq33, x2_min, color='white', alpha=1.0)
plt.fill_between(x, eq11, x2_min, color='white', alpha=1.0)

A1 = np.array([[10.2, -3.2], [-5.8, 16]])
B1 = np.array([149, 83])
intersection_point1 = np.linalg.solve(A1, B1)

A2 = np.array([[-5.8, 16], [10.3, 7.3]])
B2 = np.array([83, 234])
intersection_point2 = np.linalg.solve(A2, B2)

A3 = np.array([[10.2, -3.2], [10.3, 7.3]])
B3 = np.array([149, 234])
intersection_point3 = np.linalg.solve(A3, B3)

#         
offset1 = 4
offset2 = -15
offset3 = -10

plt.scatter(intersection_point1[0], intersection_point1[1], color='blue')
plt.text(intersection_point1[0] + offset1, intersection_point1[1] + offset1, f'({intersection_point1[0]:.2f}, {intersection_point1[1]:.2f})', fontsize=10, verticalalignment='bottom')
plt.plot([intersection_point1[0], intersection_point1[0] + offset1], [intersection_point1[1], intersection_point1[1] + offset1], color='blue', linestyle='dotted')

plt.scatter(intersection_point2[0], intersection_point2[1], color='red')
plt.text(intersection_point2[0] + offset2, intersection_point2[1] + offset2, f'({intersection_point2[0]:.2f}, {intersection_point2[1]:.2f})', fontsize=10, verticalalignment='bottom')
plt.plot([intersection_point2[0], intersection_point2[0] + offset2], [intersection_point2[1], intersection_point2[1] + offset2], color='blue', linestyle='dotted')

plt.scatter(intersection_point3[0], intersection_point3[1], color='magenta')
plt.text(intersection_point3[0] + offset3, intersection_point3[1] + offset3, f'({intersection_point3[0]:.2f}, {intersection_point3[1]:.2f})', fontsize=10, verticalalignment='bottom')
plt.plot([intersection_point3[0], intersection_point3[0] + offset3], [intersection_point3[1], intersection_point3[1] + offset3], color='blue', linestyle='dotted')



#   
plt.text(7, eq1(8), '$10.2x_1 - 3.2x_2 = 149$', fontsize=10, rotation= 38)
plt.text(-5, eq2(0), '$-5.8x_1 + 16x_2 = 83$', fontsize=10, rotation=5)
plt.text(2, eq3(8), '$10.3x_1 + 7.3x_2 = 234$', fontsize=10, rotation=-19)
#  
plt.legend()

#  
plt.grid(True)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('All lines in one figure')
plt.grid(True, color='gray', linestyle='--')

plt.show()
```
