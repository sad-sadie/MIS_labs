# Інформація про країну Джибуті

## Площа

Площа країни Джибуті: 23.2 квадратних кілометрів.

## Населення

Населення країни Джибуті: 988 000 осіб

## Державний устрій

Державний устрій Джибуті: республіка

## Видатні персоналії

* Ісмаїл Омар Гейле - Політик і державний діяч, який зіграв важливу роль у політичному житті Джибуті. Він був одним з ключових фігур у боротьбі за незалежність країни.
* Абдулкадір Качале Гуелле - Політик, який був президентом Джибуті з 1977 по 1999 рік. Під час його правління країна пережила важливі перетворення і зміни.
* Джамал Гіірі - Дипломат і політичний діяч, який відіграв важливу роль у міжнародних відносинах Джибуті. Він займав посади урядовця і представника країни в різних міжнародних організаціях.

## Пам'ятки культури та природи

1. Старе місто Жібуті (Вільйа): Це історичне місто, яке відображає багату культурну спадщину Джибуті. Воно включає в себе традиційні будівлі, базари, мечеті та інші архітектурні споруди.
2. Мечеть Хемпетіль-Альі: Це одна з найбільших і найбільш вражаючих мечетей у Джибуті. Вона має вражаючу архітектуру та є важливим центром релігійного життя для місцевого населення.
3. Наскальні малюнки Тіллілі: Розташовані в околицях Джибуті, ці давні малюнки є свідченням давньої культури і життя в регіоні. Вони представляють велике історичне значення для розуміння розвитку мистецтва і культури на території сучасного Джибуті.

## 3. Державний устрій

*Державний устрій Джибуті*: <span style="color: magenta;"> Республіка </span>


## 4. Видатні персоналі

* 4.1. **Ісмаїл Омар Гейле** - Політик і державний діяч, який зіграв важливу роль у політичному житті Джибуті. Він був одним з ключових фігур у боротьбі за незалежність країни.
* 4.2. **Абдулкадір Качале Гуелле** - Політик, який був президентом Джибуті з 1977 по 1999 рік. Під час його правління країна пережила важливі перетворення і зміни.
* 4.3. **Джамал Гіірі** - Дипломат і політичний діяч, який відіграв важливу роль у міжнародних відносинах Джибуті. Він займав посади урядовця і представника країни в різних міжнародних організаціях.
