```python
import numpy as np
```

```python
n = int(input("Введіть розмір матриці (n): "))
```

```python
%%time

def chessboard_pattern(n):
    board = np.zeros((n, n), dtype=int)
    
    board[1::2, ::2] = 1
    board[::2, 1::2] = 1
    
    return board

print("Шахова дошка розміром", n, "x", n, "з 1 та 0:")
print(chessboard_pattern(n))
```

```python
%%time
def print_board(board):
    for row in board:
        print("[", end="")
        for i in range(len(row)):
            if i == len(row) - 1:
                print(row[i], end="")
            else:
                print(row[i], end=" ")
        print("]")

def chessboard_pattern(n):
    board = []
    
    for i in range(n):
        row = []
        for j in range(n):
            if (i + j) % 2 == 0:
                row.append(0)
            else:
                row.append(1)
        board.append(row)
    
    return board



print("Шахова дошка розміром", n, "x", n, "з 0 та 1:")
board = chessboard_pattern(n)
print_board(board)
```
