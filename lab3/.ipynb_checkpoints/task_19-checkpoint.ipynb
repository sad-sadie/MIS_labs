```python
import pandas as pd
import matplotlib.pyplot as plt

csv_file_path = 'E:\\anaconda3\\anaconda files\\lab3\\NationalNames.csv'
try:
    df = pd.read_csv(csv_file_path)
except FileNotFoundError:
    print(f"  : {csv_file_path}")

births_by_year_gender = df.groupby(['Year', 'Gender'])['Count'].sum().unstack()
plt.figure(figsize=(12, 6))
plt.plot(births_by_year_gender.index, births_by_year_gender['M'], label='', color='orange')
plt.plot(births_by_year_gender.index, births_by_year_gender['F'], label='', color='blue')
plt.title('       ')
plt.xlabel('')
plt.ylabel('  ')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
