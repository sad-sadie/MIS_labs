```python
import pandas as pd
import matplotlib.pyplot as plt

csv_file_path = 'E:\\anaconda3\\anaconda files\\lab3\\NationalNames.csv'
try:
    df = pd.read_csv(csv_file_path)
except FileNotFoundError:
    print(f"Файл не знайдено: {csv_file_path}")

births_by_year_gender = df.groupby(['Year', 'Gender'])['Count'].sum().unstack()
plt.figure(figsize=(12, 6))
plt.plot(births_by_year_gender.index, births_by_year_gender['M'], label='Хлопчики', color='orange')
plt.plot(births_by_year_gender.index, births_by_year_gender['F'], label='Дівчатка', color='blue')
plt.title('Загальна кількість народжень хлопчиків та дівчаток на рік')
plt.xlabel('Рік')
plt.ylabel('Загальна кількість народжень')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
